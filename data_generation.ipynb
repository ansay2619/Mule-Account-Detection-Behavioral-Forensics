```python
import pandas as pd
import random
import numpy as np

random.seed(42)

def generate_banking_data(records=1000):
    data = []
    cities = ["Mumbai", "Delhi", "Bengaluru", "Kolkata", "Patna", "Ranchi"]
    
    for i in range(records):
        acc_id = 7000 + i
        city = random.choice(cities)
        acc_age = random.randint(1, 365)
        
        # Injected Parameter: Sleeper Logic (Idle Days)
        if i % 15 == 0: # 15th account is a 'Mule' suspect
            idle_days = random.randint(60, 200) # Long period of inactivity
            device_id = "DEV_MULE_HUB_99" # Shared across multiple accounts
            velocity = random.randint(1, 12) # Money out in < 12 mins
            txn_val = random.choice([20000, 50000, 100000]) # Round blocks
        else:
            idle_days = random.randint(0, 5) # Normal users use it immediately
            device_id = f"DEV_{random.randint(100, 800)}"
            velocity = random.randint(120, 2880) # Money stays for hours/days
            txn_val = random.randint(500, 15000)

        data.append([acc_id, city, acc_age, idle_days, device_id, velocity, txn_val])

    return pd.DataFrame(data, columns=[
        "Account_ID", "City", "Total_Age_Days", "Idle_Days", 
        "Device_ID", "Txn_Velocity_Mins", "Txn_Value_INR"
    ])

df = generate_banking_data()
df.to_csv("raw_bank_transactions.csv", index=False)
print("✅ Stage 1 Complete: 'raw_bank_transactions.csv' created.")
```
