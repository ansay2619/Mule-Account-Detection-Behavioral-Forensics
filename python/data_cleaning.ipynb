```python
import pandas as pd

df = pd.read_csv("raw_bank_transactions.csv")

def calculate_risk(row):
    score = 0
    # 1. Velocity Trap: +4 points
    if row['Txn_Velocity_Mins'] < 15: score += 4
    # 2. Shared Device Trap: +5 points
    if row['Device_ID'] == "DEV_MULE_HUB_99": score += 5
    # 3. Sleeper Trap (Idle > 60 days): +3 points
    if row['Idle_Days'] > 60: score += 3
    # 4. Round Value Check: +2 points
    if row['Txn_Value_INR'] % 10000 == 0: score += 2
    return score

df['Risk_Score'] = df.apply(calculate_risk, axis=1)

# Assign Risk Categories
def get_level(score):
    if score >= 9: return "RED ALERT (High Risk)"
    elif score >= 5: return "AMBER (Monitor)"
    else: return "GREEN (Low Risk)"

df['Risk_Level'] = df['Risk_Score'].apply(get_level)

df.to_csv("cleaned_mule_risk_assessment.csv", index=False)
```
